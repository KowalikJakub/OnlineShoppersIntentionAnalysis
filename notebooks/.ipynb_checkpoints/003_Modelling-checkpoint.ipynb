```python
import pandas as pd
import sklearn as sk
import sklearn.neural_network
import sklearn.neighbors
import sklearn.tree
import sklearn.model_selection
import scikitplot as skplt
import scikitplot.metrics
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('../data/feature_engineered_data.csv')
```

```python
X = data.drop('Revenue=True', axis=1)
y = data['Revenue=True']
X_train, X_test, y_train, y_test = \
    sk.model_selection.train_test_split(X, y, 
                                        train_size=0.6, 
                                        stratify=y,
                                        random_state=1)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
%%time
model_names = ['KNN', 'DecisionTree']
model_params_pairs = [
    (sk.neighbors.KNeighborsClassifier(), {
        'n_neighbors': list(range(10, 101, 10)),
        'weights': ['uniform', 'distance'],
        'algorithm': ['ball_tree', 'kd_tree', 'brute'],
        'p': [1,2]
    }),
    (sk.tree.DecisionTreeClassifier(), {
        'criterion': ['gini', 'entropy'],
        'splitter': ['best', 'random'],
        'max_depth': list(range(5, 101, 5)),
        'min_samples_split': list(range(20, 101, 10)),
        'min_samples_leaf': list(range(10, 51, 10))
    })
]

grid_searches = []
for est, params in model_params_pairs:
    gs = sk.model_selection.GridSearchCV(estimator=est,
                                         param_grid=params,
                                         scoring='f1',
                                         n_jobs=-1, 
                                         cv=5)
    gs.fit(X_train, y_train)
    grid_searches.append(gs)
```

```python
def specifity(y_true, y_pred):
    tn, fp, fn, tp = sk.metrics.confusion_matrix(y_true, y_pred).ravel()
    return tn / (tn+fp)

metrics = {
    'accuracy': sk.metrics.accuracy_score,
    'sensitivity': sk.metrics.recall_score,
    'specifity': specifity,
}
cls_reports = {}
for m_name, gs in zip(model_names, grid_searches):
    y_pred = gs.predict(X_test)
    report = {n: m(y_test, y_pred) for n,m in metrics.items()}
    cls_reports[m_name] = report
```

```python
pd.DataFrame(cls_reports).T.to_csv('../data/classification_results')
```

```python
skplt.metrics.plot_lift_curve(y_test, 
                                   grid_searches[0].predict_proba(X_test), 
                                   title='K-Najbliższych sąsiadów Wykres Lift',
                                   figsize=(9,6))
```
